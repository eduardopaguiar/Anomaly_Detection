```python
df["Time_s"] = df["time"].apply(lambda x: datetime.strptime(x," %H:%M:%S.%f").time())
```

```python
df
```

```python
for i in range(df.shape[0]):
    df.loc[i,'Time_s'] = ((df.loc[i,'Time_s'].hour*60 + df.loc[i,'Time_s'].minute)*60 + df.loc[i,'Time_s'].second) + df.loc[i,'Time_s'].microsecond/1000000
```

```python
df3 = df.groupby(['gra', 'func']).sum()
```

```python
df3 = df3.drop(['mode', 'time'],1)
```

```python
df3.loc[4].sort_values('Time_s', ascending=False)
```
